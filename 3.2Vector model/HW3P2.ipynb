```python
import json
import ast
from google.colab import output
```

```python
DATA_FILE_PATH = '/content/farsnews.json'
NEWS_COUNT = 1000
```

```python
# returns the first n news of JSON file

def get_n_news(n, file_path):
  News = []
  counter = 0

  for line in open(file_path, 'r', encoding='utf-8-sig'):
    News.append(json.loads(line))

    counter += 1
    if counter == n:
      break

  return News
```

```python
News = get_n_news(NEWS_COUNT + 20, DATA_FILE_PATH)
```

```python
import hazm
```

Normalization

```python
normalizer = hazm.Normalizer()

normalized_news_bodies = []
counter = 0
news_index = 0
while counter != NEWS_COUNT:
  if 'NewsBody' in News[news_index]:
    normalized_news_body = normalizer.normalize(News[news_index]['NewsBody'])
    normalized_news_bodies.append(normalized_news_body)
    counter += 1
  
  news_index += 1

print(normalized_news_bodies[1:10])
```

```python
len(normalized_news_bodies)
```

Tokenization

```python
tokenizer = hazm.WordTokenizer()

news_bodies_tokens = []

for i in range(NEWS_COUNT):
  news_body_tokens = tokenizer.tokenize(normalized_news_bodies[i])
  news_bodies_tokens.append(news_body_tokens)
```

```python
count = 0
for token_set_index in range(NEWS_COUNT):
  token_set = news_bodies_tokens[token_set_index]
  for token_index in range(len(token_set)):   
    count+=1
count
```

Stemming and posting list

```python
stemmer = hazm.Stemmer()

posting_list = {}

for token_set_index in range(NEWS_COUNT):
  token_set = news_bodies_tokens[token_set_index]

  for token_index in range(len(token_set)):
    
    token = token_set[token_index]
    
    if len(token) <= 1:
        stemmed_token = token
    else :
        stemmed_token = stemmer.stem(token)

    if len(stemmed_token) >= 3:
      stemmed_token = stemmer.stem(token)
      doc_name = 'Doc' + str(token_set_index)
      if stemmed_token in posting_list:
          d_list = posting_list[stemmed_token]
          if doc_name in d_list:
            d_list[doc_name] += 1
          else:
            d_list[doc_name] = 1
      else:
        posting_list[stemmed_token] = { doc_name : 1 }
```

```python
for item in posting_list.values():
  item['total'] = sum(item.values())
```

```python
def execute_query(query, boolean_matrix, token_index_map):
  terms = query.split('.')
  result = [1]
  for term in terms:
    if term[0] == '!':
      token = term[1:]
      if token not in token_index_map: continue
      token_index = token_index_map[token]
      vector = 1 - boolean_matrix[token_index]
    else:
      if term not in token_index_map: 
        result *= 0
        break
      token_index = token_index_map[term]
      vector = boolean_matrix[token_index]
    result *= vector
  
  return np.where(result == 1)[0]
```

```python
import collections

def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3

def union(lst1, lst2):
    final_list = lst1 + lst2
    return final_list

def calculateJaccard(doc_tokens, query_terms):
  return len(intersection(doc_tokens, query_terms))/len(union(doc_tokens,query_terms))

def executeJacQuery(query):
  doc_points={}
  query_terms = query.split()
  for doc_id, token_set in enumerate(news_bodies_tokens):
    doc_points["Doc"+str(doc_id)] = calculateJaccard(token_set, query_terms)
  counter = collections.Counter(doc_points)
  top_ten_docs = counter.most_common(10)
  return top_ten_docs
```

```python
import math

def calculate_normalized_tf(doc_tokens, query_terms):
    result = 0.0
    for q_term in query_terms:
      term_freq = doc_tokens.count(q_term)
      if term_freq != 0:
          tf_norm = 1 + math.log10(term_freq)
          result += tf_norm
    return result

def execute_normalized_tf_query(query):
  doc_points={}
  query_terms = query.split()
  for doc_id, token_set in enumerate(news_bodies_tokens):
    doc_points["Doc"+str(doc_id)] = calculate_normalized_tf(token_set, query_terms)
  counter = collections.Counter(doc_points)
  top_ten_docs = counter.most_common(10)
  return top_ten_docs
```
