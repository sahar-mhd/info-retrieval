```python
import json
import ast
from google.colab import output
```

```python
DATA_FILE_PATH = '/content/farsnews.json'
NEWS_COUNT = 1
```

```python
# returns the first n news of JSON file

def get_n_news(n, file_path):
  News = []
  counter = 0

  for line in open(file_path, 'r', encoding='utf-8-sig'):
    News.append(json.loads(line))

    counter += 1
    if counter == n:
      break

  return News
```

```python
News = get_n_news(NEWS_COUNT, DATA_FILE_PATH)
```

```python
import hazm
```

>1-2- Normalization

```python
normalizer = hazm.Normalizer()

normalized_news_bodies = []

for i in range(NEWS_COUNT):
  normalized_news_body = normalizer.normalize(News[i]['NewsBody'])
  normalized_news_bodies.append(normalized_news_body)

print(normalized_news_bodies)
```

> 1-3- tokenization


```python
tokenizer = hazm.WordTokenizer()

news_bodies_tokens = []

for i in range(NEWS_COUNT):
  news_body_tokens = tokenizer.tokenize(normalized_news_bodies[i])
  news_bodies_tokens.append(news_body_tokens)

print(news_bodies_tokens)
```

> 1-4- Stemming

```python
stemmer = hazm.Stemmer()
stopwords = hazm.stopwords_list()

posting_list = {}

for token_set_index in range(NEWS_COUNT):
  token_set = news_bodies_tokens[token_set_index]

  for token_index in range(len(token_set)):
    token = token_set[token_index]

    if not token in stopwords:
        if len(token) <= 1:
          stemmed_token = token
        else :
          stemmed_token = stemmer.stem(token)

    doc_name = 'Doc' + str(token_set_index)
    if stemmed_token in posting_list:
        d_list = posting_list[stemmed_token]

        if doc_name in d_list:
          d_list[doc_name] += 1
        else:
          d_list[doc_name] = 1
    else:
      posting_list[stemmed_token] = { doc_name : 1 }
```

```python
import dadmatools
from dadmatools.models.normalizer import Normalizer

normalizer = Normalizer(
    full_cleaning=False,
    unify_chars=True,
    refine_punc_spacing=True,
    remove_extra_space=True,
    remove_puncs=False,
    remove_html=False,
    remove_stop_word=False,
    replace_email_with="<EMAIL>",
    replace_number_with=None,
    replace_url_with="",
    replace_mobile_number_with=None,
    replace_emoji_with=None,
    replace_home_number_with=None
)
normalized_news_bodies = []

for i in range(NEWS_COUNT):
  normalized_news_body = normalizer.normalize(News[i]['NewsBody'])
  normalized_news_bodies.append(normalized_news_body)

print(normalized_news_bodies) 
```

```python
normalizer = Normalizer(full_cleaning=True)
normalized_news_bodies = []

for i in range(NEWS_COUNT):
  normalized_news_body = normalizer.normalize(News[i]['NewsBody'])
  normalized_news_bodies.append(normalized_news_body)

print(normalized_news_bodies) 
```

```python
import dadmatools.pipeline.language as language

pips = 'tok' 
nlp = language.Pipeline(pips)

# you can see the pipeline with this code
nlp.analyze_pipes(pretty=True)
news_tokens_sets = []
for i in range(NEWS_COUNT):
  doc = nlp(normalized_news_bodies[i])
  tokens = []
  dictionary = language.to_json(normalized_news_bodies[i], doc)
  for i in range(len(dictionary)):
    for j in range(len(dictionary[i])):
      tokens.append(dictionary[i][j]['text'])
  news_tokens_sets.append(tokens)
print(news_tokens_sets)
```
