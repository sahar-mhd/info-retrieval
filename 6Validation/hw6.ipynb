```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
# Initialize NLTK resources
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download('punkt')

# Define preprocessing functions
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    # Normalize
    text = text.lower()
    
    # Tokenize
    tokens = nltk.word_tokenize(text)

    return tokens
```

```python
def preprocessq(text):
    # Normalize
    text = text.lower()
    
    # Tokenize
    tokens = nltk.word_tokenize(text)
    
    # Remove stop words and lemmatize
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalnum() and token not in stop_words]
    
    return tokens
```

```python
# Process documents and calculate word frequencies
document_freq = {}
sw=False
with open("CISI.ALL", "r") as file:
    lines = file.readlines()
    doc_id = ""
    for line in lines:
        if line.startswith(".I"):
            doc_id = line.split()[1]
        elif line.startswith(".W"):
            sw=True 
            document_freq[doc_id] = {}
        elif(sw):
            if line.startswith(".X"):
                sw=False
                continue
            else:
                words = preprocess(line)
                for word in words:
                    if word in document_freq[doc_id]:
                        document_freq[doc_id][word] += 1
                    else:
                        document_freq[doc_id][word] = 1
        else:
            continue      
```

```python
# Iterate over documents and print the first 5 items of the posting list
counter = 0
for doc_id, word_freq in document_freq.items():
    print("Document:", doc_id)
    for word, frequency in word_freq.items():
        print(word, frequency)
    print()
    counter += 1
    if counter == 5:
        break
```

```python
queries = {}
query_freq = {}
with open("CISI.QRY", "r") as file:
    lines = file.readlines()
    i = 0
    while i < len(lines):
        if lines[i].startswith(".I"):
            query_id = lines[i].split()[1]
            i += 1
            if lines[i].startswith(".W"):
                i += 1
                query_text = ""
                while i < len(lines) and not lines[i].startswith(".I"):
                    query_text += lines[i].strip() + " "
                    i += 1
                queries[query_id] = query_text.strip()
        else:
            i += 1
```

```python
relevant_docs = {}
with open("CISI.REL", "r") as file:
    lines = file.readlines()
    for line in lines:
        parts = line.split()
        query_id = parts[0]
        doc_id = parts[1]
        if query_id in relevant_docs:
            relevant_docs[query_id].append(doc_id)
        else:
            relevant_docs[query_id] = [doc_id]
```

```python
from collections import Counter
import math
```

```python
def calculate_precision(retrieved_docs, relevant_docs):
    TP = len(set(retrieved_docs) & set(relevant_docs))
    FP = len(retrieved_docs) - TP
    return TP / (TP + FP)
```

```python
def calculate_recall(retrieved_docs, relevant_docs):
    TP = len(set(retrieved_docs) & set(relevant_docs))
    FN = len(relevant_docs) - TP
    if TP == 0 and FN == 0:
        return 0
    return TP / (TP + FN)
```

```python
def calculate_f_measure(precision, recall, beta):
    if precision == 0 and recall == 0:
        return 0
    return ((beta**2 + 1) * precision * recall) / ((beta**2 * precision) + recall)
```

```python
def calculate_map(retrieved_docs, relevant_docs, k):
    precision_values = []
    for i, doc_id in enumerate(retrieved_docs, start=1):
        if doc_id in relevant_docs:
            precision = calculate_precision(retrieved_docs[:i], relevant_docs)
            precision_values.append(precision)
    if precision_values:
        return sum(precision_values) / min(k, len(relevant_docs))
    return 0
```

```python
def calculate_dcg(retrieved_docs, relevant_docs):
    dcg = 0
    for i, doc_id in enumerate(retrieved_docs, start=1):
        if doc_id in relevant_docs:
            relevance = 1
            dcg += relevance / math.log2(i + 1)
    return dcg
```

```python
# Define weighting schemes
weighting_schemes = [
    ("lnc-itc", lambda tf, df, N: (1 + math.log10(tf)) * math.log10(N / df)),
    ("lnc-inc", lambda tf, df, N: 1 + math.log10(tf)),
    ("lnn-lnn", lambda tf, df, N: math.log10(1 + math.log10(tf))),
    ("ltc-ltc", lambda tf, df, N: (0.5 + 0.5 * (tf / max(document_freq[doc_id].values()))) * math.log10(N / df)),
    ("ltn-ltn", lambda tf, df, N: (0.5 + 0.5 * (tf / max(document_freq[doc_id].values())))),
    ("bnn-bnn", lambda tf, df, N: 1 if tf > 0 else 0)
]
```

```python
results = []
df=0
for weighting_scheme, _ in weighting_schemes:
    for remove_stop_words in [True, False]:
        for stemming in [True, False]:
            # Apply preprocessing if necessary
            if remove_stop_words:
                document_freq_processed = {
                    doc_id: {word: freq for word, freq in freqs.items() if word not in stop_words}
                    for doc_id, freqs in document_freq.items()
                }
                queries_processed = {
                    query_id: " ".join(preprocess(query_text))
                    for query_id, query_text in queries.items()
                }
                if stemming:
                    stemmer = PorterStemmer()
                    document_freq_processed = {
                        doc_id: {stemmer.stem(word): freq for word, freq in freqs.items()}
                        for doc_id, freqs in document_freq.items()
                    }
            else:
                document_freq_processed = document_freq
                queries_processed = queries

            for query_id, query_text in queries_processed.items():
                # Preprocess query
                query_terms = preprocessq(query_text)
                
                # Retrieve relevant documents using the chosen weighting scheme
                retrieval_scores = Counter()

                for doc_id, freqs in document_freq_processed.items():
                    score = 0
                    for term in query_terms:
                        if term in freqs:
                            tf = freqs[term]  # TF (term frequency) for the current term in the current document
                            df = 0
                            for freqs in document_freq_processed.values():
                                if term in freqs:
                                    df += 1

                            weighting_scheme_func = next(func for scheme_name, func in weighting_schemes if scheme_name == weighting_scheme)
                            score += weighting_scheme_func(tf, df, len(document_freq_processed))
                    retrieval_scores[doc_id] = score

                
                # Retrieve the top 20 documents
                retrieved_docs = [doc_id for doc_id, _ in retrieval_scores.most_common(20)]
                
                # Calculate evaluation criteria
                precision = calculate_precision(retrieved_docs, relevant_docs.get(query_id, []))
                recall = calculate_recall(retrieved_docs, relevant_docs.get(query_id, []))
                f_measure = calculate_f_measure(precision, recall, 1)
                map_ = calculate_map(retrieved_docs, relevant_docs.get(query_id, []), 20)
                dcg = calculate_dcg(retrieved_docs, relevant_docs.get(query_id, []))
                
                # Add the results to the list
                results.append((remove_stop_words, stemming, weighting_scheme, precision, recall, f_measure, map_, dcg))
```

```python
# Print the results table
print("Remove Stop Words\tStemming\tWeighting Scheme\tPrecision\tRecall\tF-Measure\tMAP\tDCG")
for result in results:
    remove_stop_words, stemming, weighting_scheme, precision, recall, f_measure, map_, dcg = result
    print(f"{remove_stop_words}\t\t\t{stemming}\t\t\t{weighting_scheme}\t{precision:.4f}\t\t{recall:.4f}\t{f_measure:.4f}\t\t{map_:.4f}\t{dcg:.4f}")
```

```python
# Calculate overall accuracy by averaging the results
accuracy = sum(result[3] for result in results) / len(results)
print(f"\nOverall Accuracy: {accuracy:.4f}")
```
